```python
import pandas as pd
import numpy as np
raw_mail_data = pd.read_csv("/content/mail_data.csv")
raw_mail_data.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

```python
raw_mail_data.isnull().sum() #checking wether null value
```

```python
df = raw_mail_data.where((pd.notnull(raw_mail_data)),'') #replace null value wiht a null string
df.isnull().sum()
```

```python
df.shape
```

LABEL ENCODING

```python
df['Category']= df['Category'].map({'spam':0,'ham':1})
df.head(10)
```

Using Loc Function

```python
df.loc[df['Category']=='spam','Category',]= 0
df.loc[df['Category']=='ham','Category',]= 1
df.head(10)
```

```python
X= df['Message']
Y=df['Category'] #dependent or independent variable
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
Y_test.shape
```

```python
feature_extraction=TfidfVectorizer(min_df = 1,stop_words="english",binary=True)
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)
```

```python
df.info()
```

```python
Y_train=Y_train.astype("int")
Y_test=Y_test.astype("int")
```

```python
X_train
```

```python
print(X_train_features)
```

```python
print(X_test_features)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_features,Y_train)
```

```python
prediction_train_data=model.predict(X_train_features)
training_data_accuracy=accuracy_score(Y_train,prediction_train_data)
```

```python
training_data_accuracy
```

```python
prediction_test_data=model.predict(X_test_features)
testing_data_accuracy=accuracy_score(Y_test,prediction_test_data)
```

```python
testing_data_accuracy
```

```python
input_user_mail=["WINNER!! As a valued network customer you have been selected to receivea £900 prize reward! To claim call 09061701461. Claim code KL341. Valid 12 hours only."]
input_data_features = feature_extraction.transform(input_user_mail)
prediction= model.predict(input_data_features)

if prediction[0]==1:
  print('Ham mail')
else:
  print('Spam mail')
```
